```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
v_koh = np.arange(0, 8.5, 0.5)
pH1 = np.array([6.843,8.456,8.873,9.068,9.234,9.349,9.466,9.567,9.640,9.744,9.827,9.915,10,10.087,10.169,10.263,10.367])

d = {
    'V KOH': v_koh,
    'pH': pH1,
}
d = pd.DataFrame(d)
d
```

```python
H = np.power(10, -pH1)
K = (v_koh * 0.5) / (110 + v_koh)
OH = (pow(10, -14) / H)
A = K
At = (10 * 0.4) / 110  
HA = At - A
HA2 = At - (K + H - OH)


pka = pH1 + np.log10(HA/(K + H - OH))
ka = pow(10, -pka)

data1 = {
    'V KOH': v_koh,
    'pH': pH1,
    '[H]': H,
    '[K]': K,
    '[OH]': OH,
    '[A]': A,
    '[HA]': HA,
    'pKa': pka,
    'Ka': ka,
}

data1 = pd.DataFrame(data1)
data1
```

```python
Ka = np.mean(ka)
```

```python
plt.plot(v_koh, pH1)
plt.title('Titration plot of Glycinate')
plt.xlabel('Volume of KOH')
plt.ylabel('pH')
plt.show()
```

```python
H = np.power(10, -pH2)
OH = (pow(10, -14) / H)
CH = (10 * 0.1) / (200 + v)
A = (Ka * (CH + OH - H)) / H
Mt = 1 / (200 + v)

At = (10 * 0.4) / 210

n = (At - ((1 + Ka/H) * (CH + OH - H))) / Mt
n
```
