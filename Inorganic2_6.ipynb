```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_en = np.array([0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 0])
l_530 = np.array([0.130, 0.280, 0.388, 0.493, 0.426, 0.328, 0.185, 0.039])
l_545 = np.array([0.198, 0.362, 0.483, 0.543, 0.425, 0.334, 0.186, 0.046])
l_578 = np.array([0.381, 0.573, 0.563, 0.459, 0.320,0.242,0.137,0.095])
l_622 = np.array([0.521, 0.545, 0.453, 0.234, 0.178, 0.111, 0.065, 0.279])
l_640 = np.array([0.498, 0.492, 0.385, 0.174, 0.148, 0.084, 0.049, 0.379])

d = {
    'x_en': x_en,
    '530': l_530,
    '545': l_545,
    '578': l_578,
    '622': l_622,
    '640': l_640,
}

d = pd.DataFrame(d)
d.index = [i+1 for i in range(8)]
d
```

```python
y_530 = l_530[:7] - (l_530[7] * (1 - x_en[:7]))
y_545 = l_545[:7] - (l_545[7] * (1 - x_en[:7]))
y_578 = l_578[:7] - (l_578[7] * (1 - x_en[:7]))
y_622 = l_622[:7] - (l_622[7] * (1 - x_en[:7]))
y_640 = l_640[:7] - (l_640[7] * (1 - x_en[:7]))
```

```python
plt.figure(figsize= (20, 3))

plt.subplot(151)
plt.plot(x_en[:7], y_530)
plt.xticks(np.arange(0, 1, 0.1))

plt.subplot(152)
plt.plot(x_en[:7], y_545)


plt.subplot(153)
plt.plot(x_en[:7], y_578)

plt.subplot(154)
plt.plot(x_en[:7], y_622)

plt.subplot(155)
plt.plot(x_en[:7], y_640)

plt.show()
```

```python
plt.figure(figsize= (4,4))
plt.plot(x_en[:7], y_530)
plt.xticks(np.arange(0, 1, 0.1))
plt.xlabel('X_en')
plt.ylabel('Y')
plt.title('530nm')
plt.show()
```
